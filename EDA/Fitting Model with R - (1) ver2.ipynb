## 1. Fitting Ridge Model

## 5. Fitting Regression Tree Model

```r
set.seed(2021)
train<-sample(1:nrow(x),nrow(x)*0.8) 
test<-(-train)
y.test<-y[test]
```

```r
x <- as.data.frame(x)
accident_cnt <- y
```

```r
total_sub_df <- cbind(accident_cnt,x)
total_sub_df
```

```r
library(rpart)
```

```r
my.control = rpart.control(xval=10,cp=0,minsplit=ceiling(nrow(x[train,])*0.05))
```

```r
fit.tree=rpart(accident_cnt~.,data=total_sub_df[train,],method='anova',control=my.control) # anova : 회귀나무 / class : 분류나무
```

```r
which.min(fit.tree$cp[,4])
```

```r
fit.tree$cp[2,]
```

```r
fit.tree$cp[2,1] # best CP
```

```r
ii = which.min(fit.tree$cp[,4])
fit.prune.tree = prune(fit.tree,cp=fit.tree$cp[ii,1]) # 가지치기
```

```r
yhat.tree = predict(fit.prune.tree,newdata=total_sub_df[test,],type='vector')
mean((total_sub_df[test,]$accident_cnt - yhat.tree)^2)
```

- Regression Tree Test MSE : 0.99607

### Validation Set Approach
- train 0.8 / test 0.2

```r
set.seed(2021)
train = sort(sample(1:dim(total_sub)[1],dim(total_sub)[1]*0.8))       # 8대2 비율로 train set의 index를 random sampling
test = setdiff(1:dim(total_sub)[1],train)

y_test = total_sub$accident_cnt[test]      # test set
x_test = total_sub[test,-1]

lm.fit = lm(accident_cnt ~ 생산가능인구_count_100 + 주정차_count_300 + 학원수_count_300 + 어린이집_count + 
    정류장_count_100 + 정류장_count_200 + 총인구_count_100 + 학원수_count_100,data=total_sub,subset=train)
yhat = predict(lm.fit,x_test)         # x_test를 새로운 data로 predict
mean((y_test-yhat)^2)
```

- Validation set Approach(8:2) Test MSE : 1.0499
